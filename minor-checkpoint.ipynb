Q1)Which movie had the most profit during 2000-2005?

```python
df_years=df[(df['release_year' ]<2004) & (df['release_year' ] >=2000)]
```

```python
df_years.head()
```

```python

df_years['profit']=df_years['revenue']-df_years['budget']
```

```python
df_years.head()
```

```python
df_years[df_years['profit']==df_years.profit.max()]
```
